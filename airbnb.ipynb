## Fetching the Amenities details

```python
amenities= []
for i in coll.find({},{"_id":1, "amenities":1}): # Fetching id and amenities columns from the collection
    amenities.append(i) # Appending all the id's and amenities details into empty_list (amenities)
```

```python
df_amenities= pd.DataFrame(amenities)
```

```python
# Sorting names of amenities in ascending order
def sort_amenities(x):
    a = x
    a.sort()
    return a
```

```python
df_amenities["amenities"]= df_amenities["amenities"].apply(lambda x: sort_amenities(x)) 
```
